二、（15分）某市的自来水执行阶梯水价（单位：$m^3$），其收费标准为：

- 居民用水：26$m^3$及以下，每$m^3$ 2.97元，(26, 34]之间的部分，每$m^3$ 3.93元，34$m^3$以上的部分，每$m^3$ 6.79元；
- 非居民用水：每$m^3$ 4.17元；
- 特种用水：每$m^3$ 6.57元.

编写一个函数water_fee(ton, cls)，实现这一收费方案，对输入的当月用水量ton和用水类型cls（0为居民，1为非居民，2为特种，默认为居民），返回其对应费用。

```python
# 以下为函数定义
def water_fee(ton, cls=0):
    if cls==0:
        if ton<=26:
            cost = ton*2.97
        elif ton<=34:
            cost = 26*2.97 + (ton-26)*3.93
        else:
            cost = 26*2.97 + 8*3.93 + (ton-34)*6.79
    elif cls==1:
        cost = ton*4.17
    elif cls==2:
        cost = ton*6.57
    return cost

# 以下为函数调用
print(water_fee(25))
print(water_fee(37,0))
print(water_fee(47,1))
print(water_fee(73,2))
```

三、（10分）如果一个自然数能整除其各位数字之和，则称为“哈沙德数”。如48能整除4+8，36能整除3+6，它们都是哈沙德数。求[100,199]上的所有哈沙德数。

```python
hashade = []
for k in range(100, 200):
    d = [int(x) for x in str(k)]  # 获得各位数字
    if k%sum(d)==0:
        hashade.append(k)

print('[100,199]上所有的哈沙德数有：\n',hashade)
```

四、(20分) 分别绘制下列图形：<br>
1. 某公司去年5个部门的营收分别为32.3, 40.1, 24.7, 53.9, 29.6亿元，分左右子图，分别用条形图和饼图表现这些数据；条形图横坐标为['dp01','dp02','dp03','dp04','dp05']，标题为'income'；饼图上应显示比例，标题为'ratio'。<br>
2. 曲面$z = -\dfrac{x^3}6 + \dfrac{y^2}2$在区域 $x,y\in[-2,2]$上的图形，$x,y$的步长取为0.1。提示：在`subplot()`中添加参数`projection='3d'`即可创建3维坐标系ax，用`ax.plot_surface()`可绘制曲面图。 

```python
# 条形图和饼图
import numpy as np
import matplotlib.pyplot as plt

cost = [32.3, 40.1, 24.7, 53.9, 29.6]
labels = ['dp01','dp02','dp03','dp04','dp05']
plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.bar(labels, cost)
plt.title('income')
plt.subplot(122)
plt.pie(cost, labels=labels, autopct="%.2f%%")
plt.title('ratio')
plt.show()
```

```python
# 曲面图
import numpy as np

t = np.linspace(-2, 2, 41)  # t=np.arange(-2,2.1,0.1)
x, y = np.meshgrid(t, t)
z = -x**3 / 6 + y*y / 2
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='jet')
plt.show()
```
