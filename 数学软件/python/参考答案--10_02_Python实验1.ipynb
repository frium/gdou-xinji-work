1.使用循环计算下列级数之和。
    
$(1)\displaystyle \sum_{n=1}^{100}\frac1{4^n}$ &emsp; &emsp; &emsp; &emsp; $(2)\displaystyle \sum_{n=1}^{100}(-1)^{n+1}\frac1n$

```python
s1 = 0  # 累加和
d = 1  # 通项
for n in range(1,101): # 以下不用乘方，以减少运算量
    d /= 4
    s1 += d
print(s1)
```

```python
s2 = 0  # 累加和
c = 1  # 正负1
for n in range(1,101):
    s2 += c/n
    c = -c
print(s2)
```

2. 某商场促销活动，总价低于100元内不打折，[100,200)打9折，[200,500)打8.5折，[500,1000)打8折，1000以上打7.5折。现有顾客数人，他们购买商品总价分别为[355, 704, 256, 883, 1521, 197, 556]，请给出打折后的价格，输出格式为："总价为s的商品，打折后为t"，精确到小数点后2位。注：round(t,n) 表示对t四舍五入到小数点后n位。

```python
for s in [355, 704, 256, 883, 1521, 197, 556]:
    if s <100:
        r = 1;
    elif s <200:
        r = 0.9
    elif s < 500:
        r = 0.85
    elif s < 1000:
        r = 0.8
    else:
        r = 0.75
    print("总价",s,"的商品，打折后为",round(s*r,2))
```

3. 某人贷款50万，月利率为0.31%，每月归还4000元，则需要多少个月才能还清，他总共还了多少钱(精确到小数点后2位)?

```python
loan = 500000 # 贷款总量
r = 0.31/100  # 月利率
repay = 4000      # 月供
month = 0     # 还款所需时间(月)
while loan>0:
    loan = loan*(1+r)-repay  # 尚欠银行贷款
    month += 1
total = repay*month + loan # 因最后一月可能不需还4000，故循环结束时余额loan为负数
print('需要'+str(month)+'月才能还清')
print('总共还了', round(total,2), '元')
```

4. Syracuse序列的生成从一个自然数开始，重复应用以下迭代关系，直至等于1：$$syr(x)=\begin{cases}x/2, \quad x为偶数,\\3x+1,\quad x为奇数.\end{cases}$$ 
&emsp;例如，从5开始的Syracuse序列是5,16,8,4,2,1。编写程序，输入一个自然数，输出以此为起始值的Syracuse序列。

```python
s= int(input('请输入Syracuse序列的第一个数：'))
print(s, '为起始值的Syracuse序列为：')
while s != 1:
    print(int(s), end=' ')
    if s%2:  # s%2等于1时表示奇数
        s = 3*s+1
    else:
        s = s/2
print(1)
```

5. 若一个数等于它的各个真因子之和，则称该数为完数，如6=1+2+3，所以6是完数。编写程序求[1,500]之间的全部完数。

```python
for n in range(1,501):
    s = 0  # n的真因子之和
    for f in range(1, n//2+1):
        if n%f==0:
            s=s+f
    if s==n:
        print(n,'是完数')
```

6. 使用辗转相除法求2个任意正整数的最大公约数和最小公倍数. 依据是：m与n的最大公约数也是n与m%n的最大公约数.

```python
m=int(input("请输入第一个正整数，m="))
n=int(input("请输入第二个正整数，n="))
L=m*n  #记录m,n的乘积
while m%n: # 余数不等于0
    m,n = n,m%n   # 利用序列赋值便于交换数值
print('最大公约数为：', n)
print('最小公倍数为：', L//n) # 两数之积等于其最大公约数与最小公倍数之积
```

7. 编写程序，用户从键盘输入一个正整数，对其进行质因子分解。

```python
# 对正整数做质因子分解
n = int(input('请输入一个正整数: n='))
print(n, '的质因子分解为：')
k = 2
while n>1:  # n>1表示分解未结束
    while n%k==0: # 若n能被k整除，则继续整除k
        print(k, end=' ')
        n=n//k 
    k+=1  # 循环 "while n%k==0"结束，表示n不能整除k，则k自增1
```

8. （选做题）编写程序，实现一个猜数游戏。由程序产生一个[10,100]内的随机整数，然后要游戏者猜数的大小（即要求游戏者输入所猜的整数）。程序判断游戏者猜的数与设定的数的大小，并给出“大了”或“小了”的判断，再次要求游戏者猜一个数，然后程序再判断，直至数字与设定的数相同，则给出“正确，游戏通关！”的信息。

```python
from random import randint
num = randint(10,100)
print('猜数游戏开始！')
k = int(input('请输入[10,100]上的一个数字：'))
while num-k:  # 表示num!=k
    if k<num:
        k = int(input('小了！再猜: '))
    else:
        k = int(input('大了！再猜: '))
else:
    print('正确，游戏通关！')
```

9. （选做题）编写程序，输出九九乘法口诀表，要求用下三角形式显示，且用汉字表现乘法口诀：如三三得六，三四十二，...

```python
# 输出中文版的九九乘法口诀表

dig = '零一二三四五六七八九十'
for i in range(1, 10):  # 乘数从1到9中取值
    for j in range(1, i + 1):
        if i*j<10:  # 乘积为个位数，用4个汉字表示
            print('%s%s得%-3s'%(dig[j], dig[i], dig[j*i]), end=' ')
        elif 10<i*j<20:  # 乘积在(10, 20)内，用4个汉字表示
            k = i*j%10
            print('%s%s%s%-3s'%(dig[j],dig[i],dig[10],dig[k]), end=' ')
        elif i*j%10==0:  # 乘积为10的倍数，用4个汉字表示
            k = i*j//10
            print('%s%s%s%-3s'%(dig[j],dig[i],dig[k],dig[10]), end=' ')
        else:   # 其他，用5个汉字表示
            k1 = i*j//10
            k2 = i*j%10
            print('%s%s%s十%s'%(dig[j],dig[i],dig[k1],dig[k2]), end=' ')        
    print()  # 回车，另起一行
```
