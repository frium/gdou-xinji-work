2. 用list()函数能分离字符串的特点，求所有的水仙花数：即一个3位数的各位数字的立方和等于该数本身。

```python
for k in range(100,1000):
    nums = [int(i) for i in list(str(k))] 
    sum_nums = sum([x**3 for x in nums])
    if k==sum_nums:
        print(k,'是水仙花数')
```

```python
# 因字符串本身可以作为迭代对象，所以上面代码的第2行可以简化，即不需要list函数
for k in range(100,1000):
    nums = [int(i) for i in str(k)]
    sum_nums = sum([x**3 for x in nums])
    if k==sum_nums:
        print(k,'是水仙花数')
```

4. 回文数是指一个数字从左边读和从右边读的结果是一模一样的,比如12321。从键盘任意输入一个正整数，判断其是否为回文数，使用列表完成该问题。

```python
# 以下纯用循环完成判断
n = int(input('请输入一个正整数，n='))
nums = []
n1 = n  # 记录n的值，以备后续使用
while n>0:
    k = n%10  # 取n的末位数
    nums.append(k)  # 将n的各位数从后往前存入nums中
    n = (n-k)/10  # n中去掉末位数
flag = True
for i in range(len(nums)):
    if nums[i]!=nums[-i-1]:  # 前后对比
        flag = False
        break
if flag:
    print(n1, '是回文数')
else:
    print(n1, '不是回文数')    
```

5. 矩形法是数值方法求定积分的最基本方法。即 $\int_a^b f(x) dx \approx \sum\limits_{k=0}^{n-1} f(x_k)\cdot\frac{b-a}n$，这里 $n$ 为 区间$[a,b]$ 的等分数。试用该方法求 $\int_0^{\pi}\sin xdx$的近似值，取 $n=100.$ 

```python
# 以下用循环完成
from math import pi, sin

a,b = 0, pi
n = 100
h = (b-a)/n
S = 0
for k in range(n):
    S += sin(a+h*k)
S *= h
print('定积分的近似值为:', S)
```

```python
# 以下用列表推导式
from math import pi, sin

a,b = 0, pi
n = 100
h = (b-a)/n
sinx = [sin(a+k*h) for k in range(n)]
S = sum(sinx)*h
print('定积分的近似值为:', S)
```

6. 若一个数等于它的各个真因子之和，则称该数为完数，如6=1+2+3，所以6是完数。编写程序求[1,500]之间的全部完数，并按如下格式输出：<br>
    6 = 1+2+3  是完数

```python
# 以下的程序同时输出完数的所有真因子，这里用列表保存真因子

for n in range(1,501):
    factors = []
    for f in range(1,n//2+1):
        if n%f==0:
            factors.append(f)    
    if sum(factors)==n:
        print(n,end=' = ')
        for f in factors[:-1]:
            print(f, end='+')
        print(factors[-1],'是完数') # 最后一个单独输出
```

```python
# 使用列表推导式求真因子

for n in range(1,501):
    factors = [f for f in range(1,n//2+1) if n%f==0]
    if sum(factors)==n:
        print(n,end=' = ')
        for f in factors[:-1]:
            print(f, end='+')
        print(factors[-1],'是完数') # 最后一个单独输出
```

```python
# 如果只是求全部的完数，则用嵌套的列表推导式比较简单

[n for n in range(1,501) if n==sum([f for f in range(1,n//2+1) if n%f==0])]
```

7. （选做题）使用random模块的randint函数产生[1,10]上的100个随机整数，放到一个列表中，并输出各不同的数字及其出现次数。

```python
# 以下使用循环记录列表中的不同数字的出现频数
from random import randint

rands = []
for k in range(100): # 产生100个随机数
    rands.append(randint(1,10))
print(rands)
nums = [0*x for x in range(10)] # 记录频数，初始化为0
for k in rands:
    nums[k-1]+=1
for i in range(1,11):
    print('数字%d在列表中出现了%d次'%(i, nums[i-1])) 
```

```python
# 以下利用列表的方法 .count()对列表元素计数
from random import randint

rands = [randint(1,10) for k in range(100)] # 产生100个随机数
print(rands)
for i in range(1,11):
    print('数字%d在列表中出现了%d次'%(i, rands.count(i)))
```

8. （选做题）今天是11月12日，星期二。编写程序，对本月的任一日期，输出对应的星期。

```python
days = '一二三四五六日'
today = (12, '二')
r0 = days.index(today[1])- today[0]%7  # 计算初始日(即第0天)对应的星期序号
for date in range(1,31):
    r = (date+r0)%7
    print(date,'日：','星期'+days[r])
```

```python
days = '一二三四五六日'
for date in range(1,31):
    ind = (date-(12-1))%7  # 12日对应到days中的序号1（星期二）
    print(date, '日：', '星期'+days[ind])
```

9. （选做题）16位银联信用卡的最后一位为校验码，采用的是LUHN算法，也叫模10算法。其计算方法为：<br>
   （1）从卡号最后一位数字开始，逆向将奇数位(1、3、5等等)相加。<br>
   （2）从卡号最后一位数字开始，逆向将偶数位数字，先乘以2，再将其各位数求和。<br>
   （3）将奇数位总和加上偶数位总和，结果能被10整除。<br>
&emsp;请你编程产生10个信用卡号：卡号的前6位固定为625965，接下来的9位从0~9中随机取值，然后依LUHN算法产生最后一位检验码。输出这10个16位卡号。

```python
from random import randint

for i in range(10):
    s1 = [6,2,5,9,6,5]
    s2 = [randint(0, 9) for k in range(9)]
    s = s1 + s2  # 前15位
    p1 = sum(s[-2::-2]) # 16位卡号（去掉末位）的逆向奇数位相加
    p2 = 2*sum(s[-1::-2]) # 16位卡号的逆向偶数位相加
    t = (-p1-p2) % 10 # 使 p1+p2+t能被10整除
    cardid = ''.join(map(str, s1 + s2 + [t])) # 卡号
    print(cardid)
```
