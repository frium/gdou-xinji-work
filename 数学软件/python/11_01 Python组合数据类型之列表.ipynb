<font face='kaiti' size=5 style='line-height:1.5'>

&emsp; 实际问题中往往需同时处理多个相关数据，这需要有效地整合数据。Python中内置了几种组合数据类型，能够将多个同类型或不同类型的数据有效地组织起来并统一表示。

<font face='kaiti' size=5 style='line-height:1.5'>

&emsp; 根据数据间的关系，组合数据类型可以分为3类：<font color=#FF0000>序列类型、映射类型和集合类型</font>。

<font face='kaiti' size=5 style='line-height:1.8'>

&emsp; 序列类型是一维向量，元素之间存在先后关系，通过整数序号作索引号访问。Python中的序列类型主要有列表、元组、字符串。<br>
&emsp; 映射类型是“键-值”数据项的组合，每个元素是一个键值对，即(key, value)，元素间是无序的。键表示一个属性，值是属性的内容。在Python中，映射类型主要以字典(dict)体现。<br> 
&emsp; 集合类型和数学中集合的概念一致，是包含0个或多个数据项的无序组合。集合中的元素不可重复，元素类型只能是不可变类型。

<font face='kaiti' size=5 style='line-height:1.8'>

&emsp;根据数据的内存管理方式，Python的数据类型可分别<font color=#FF0000>可变数据类型和不可变数据类型</font>。其根本的区别在于当该数据类型的对应变量的值发生了改变，它对应的内存地址是否发生改变。内存地址随值改变，则为不可变数据类型，若内存地址不随值改变，则为可变数据类型。<br>
&emsp;数值数据、字符串、元组是不可变数据类型，而列表、字典、集合是可变数据类型。

## 列表

<font face='kaiti' size=5 style='line-height:1.5'>
    
- 列表（list）是Python中最重要的数据结构之一，是包含若干元素的有序连续内存空间。  
- 在形式上，列表的所有元素放在一对方括号[]中，相邻元素之间使用逗号分隔。如果只有一对方括号而没有任何元素则表示空列表。  
- 在Python中，同一个列表中元素的数据类型可以各不相同，可以同时包含整数、实数、字符串等基本类型的元素，也可以包含列表、元组、字典、集合、函数以及其他任意对象。

<font face='kaiti' size=5 style='line-height:1.5'>

&emsp; 示例：  
```python
[10, 20, 30, 40]
['crunchy frog', 'ram bladder', 'lark vomit']
['spam', 2.0, 5, [10, 20]]
[{3}, {5:6}, (1, 2, 3)]
```

### 列表元素的双向索引和访问

<font face='kaiti' size=5 style='line-height:1.8'>
    
&emsp; 创建列表之后，可以使用整数作为索引号来访问其中的元素，<font color=#ff0000>索引从0开始</font>。其中0表示第1个元素，1表示第2个元素，2表示第3个元素，以此类推。<br>

````python
x = list('Python')
x[0], x[-1]
````

````python
y = range(3,10)  # range对象也支持双向索引
y[2], y[-2]
````

### 切片操作的强大功能

<font face='kaiti' size=5 style='line-height:1.8'>
    
&emsp; 切片，又称区间索引，用于同时访问多个元素，是Python序列的重要操作之一，类似于MATLAB中的冒号表达式。除了适用列表之外，还适用于元组、字符串、range对象。<br>

<font face='kaiti' size=5 style='line-height:1.5'>
    
&emsp; 在形式上，切片使用2个冒号分隔的3个数字来完成，如下所示：
```python
[start:end:step]   # 注意3个数字的顺序与MATLAB中有所不同
```

<font face='kaiti' size=5 style='line-height:1.5'>

- 第一个数字start表示切片开始位置，默认为0；  
- 第二个数字end表示切片截止（<font color=#ff0000><u>但不包含</u></font>）位置（默认为列表长度）；  
- 第三个数字step表示切片的步长（默认为1）。  
- 当start为0时可以省略，当end为列表长度时可以省略，当step为1时可以省略，省略步长时还可以同时省略最后一个冒号。 
- start和end支持双向索引，即也可以用负整数。
- 当step为负整数时，表示反向切片，这时start应该在end的右侧才行。

<font face='kaiti' size=5 style='line-height:1.8'>

列表的切片操作具有最强大的功能。可以通过切片返回列表的部分元素形成新列表，还可以通过切片修改、删除、或增加元素。

<font face='kaiti' size=5 style='line-height:1.8'>
    
（1）使用切片获取列表部分元素<br>
&emsp; 使用切片可以返回列表中部分元素组成的新列表。

````python
aList = [3, 4, 5, 6, 7, 9, 11, 13, 15, 17]
aList[1:7:2]   # 以步长2返回序号从1到7的元素，不包括序号7的元素
````

````python
aList[3:6]     #指定切片的开始和结束位置，步长为1
````

````python
aList[3:]     # 返回序号3后的所有元素
````

````python
aList = [3, 4, 5, 6, 7, 9, 11, 13, 15, 17]
aList[1::2]    #从序号1开始隔一个取一个
````

````python
aList[::2]     # 从最开始隔一个取一个
````

````python
aList[:-1]    # 返回最后一个元素前的所有元素
````

````python
aList[-1]
````

````python
aList[::-1]     #返回包含原列表中所有元素的逆序列表
````

````python
aList = [3, 4, 5, 6, 7, 9, 11, 13, 15, 17]
aList[:]     #返回原列表的浅复制，也可用aList[::]
````

````python
aList[3:-10:-1]     # 序号3在序号-10的右侧，-1表示反向切片
````

````python
aList[3:-5]       #序号3在序号-5的左侧，正向切片
````

<font face='kaiti' size=5 style='line-height:1.8'>

&emsp;与索引不同，<font color=#ff0000>切片操作不会因为下标越界而抛出异常</font>，而是简单地在列表尾部截断或者返回一个空列表，代码具有更强的健壮性。

````python
aList = [3, 4, 5, 6, 7, 9, 11, 13, 15, 17]
aList[0:100]     #切片结束位置越界，从列表尾部截断
````

````python
aList[100:]      #切片开始位置大于列表长度时，返回空列表
````

````python
aList[-15:3]     # 序号`-15`越界，进行必要的截断处理
````

````python
aList[100]       #非切片，索引越界，抛出异常
````

<font face='kaiti' size=5 style='line-height:1.8'>
    
（2）使用切片为列表增加元素  
&emsp;可以使用切片操作在列表任意位置插入新元素，不影响列表对象的内存地址，属于<font color=#ff0000>原地操作</font>，不需用赋值语句返回值。<font color=#ff0000>注意：因为列表的切片还是列表，所以等号"="右边也应是列表形式</font>。

````python
aList = [3, 5, 7]
aList[1:1] = [4]               #在列表的第1个元素后插入元素
aList
````

````python
aList[1:1]  # 空列表
````

````python
aList[len(aList):] = [9]       #在列表尾部增加元素，也可用 aList[-1:-1] = [9]
aList
````

````python
aList[:0] = [1, 2]             #在列表头部插入多个元素，相当于aList[0:0]=[1,2]
aList
````

<font face='kaiti' size=5 style='line-height:1.5'>（3）使用切片替换和修改列表中的元素

````python
aList = [3, 5, 7, 9]
aList[-1:]=[8]                 # 替换最后一个元素
aList
````

````python
aList[:3] = [1, 2, 3, 4]     #切片连续，等号两边的列表长度可以不相等
aList
````

````python
aList[3:] = [4, 5, 6]         #切片连续，等号两边的列表长度可以不相等
aList
````

````python
aList[::2] = [0,0,0]         #隔一个修改一个，等号两边的列表长度需相等
aList
````

<font face='kaiti' size=5 style='line-height:1.5'>
    
（4）使用切片删除列表中的元素：可将切片赋值为空列表，或使用`del`命令。

````python
# 连续切片赋值为空列表
aList = [3, 5, 7, 9]
aList[:2] = []                  #删除列表中前2个元素
aList
````

````python
# del命令结合切片删除部分元素
aList = [3, 5, 7, 9, 11]
del aList[:3]                   #切片元素连续
aList
````

````python
# del命令删除元素时，切片元素可以不连续。
aList = [3, 5, 7, 9, 11]
del aList[::2]                  #切片元素不连续，隔一个删一个
aList
````

<font face='kaiti' size=5 style='line-height:1.5'>

- 当列表增加或删除元素时，列表对象<span class="mark">自动进行内存的扩展或收缩</span>，从而保证相邻元素之间没有缝隙。这一内存自动管理功能可以大幅度减少程序员的负担。但插入和删除非尾部元素时涉及到列表中大量元素的移动，会严重影响效率。在非尾部位置插入和删除元素时会改变该位置后面的元素在列表中的索引，这对于某些操作可能会导致意外的错误结果。除非确实有必要，否则应尽量从列表尾部进行元素的追加与删除操作。

<font face='kaiti' size=5 style='line-height:1.8'>

- 列表是可变数据类型。当仅仅改变列表中的元素，而非直接对列表本身重新赋值时，列表的内存地址不变。这样的操作又称为<font color=#FF0000>原地操作</font>。

````python
# 以下操作均不改变列表的内存地址
alist = [1,2,3,4,5]
print(id(alist))
alist[0] = 0
print(id(alist))
del alist[2]
print(id(alist))
alist[1:1]=[1]
print(id(alist))
alist
````

### 列表对象支持的运算符

<font size=4>

$$\begin{array}{c|l|l}\hline
运算符 & 形式 & 含义\\ \hline
 + & {\rm aList + bList} & 两个列表的元素合并成一个列表, 返回新列表 \\
 += & {\rm aList\; += bList} & 在{\rm aList}中添加{\rm bList}的元素，不返回新列表，属原地操作\\
 * & {\rm aList * n} & 将列表{\rm aList}中的元素复制{\rm n}份，返回新列表 \\
 *= & {\rm aList *= n} & 将{\rm aList}中的元素复制{\rm n}份，不返回新列表，属原地操作 \\
 {\rm in}& {\rm n\; in\; aList} & 测试{\rm n}是否是{\rm aList}的元素\\ \hline
\end{array}$$

<font face='kaiti' size=5 style='line-height:1.5'>
    
&emsp; 加法运算符"+"可以实现列表增加元素的目的，但不属于原地操作（内存地址改变），而是返回新列表，涉及大量元素的复制，效率低。使用复合赋值运算符"+="实现列表追加元素时属于原地操作（内存地址不变），与append()方法一样高效。 

````python
x = [1, 2, 3]
print(id(x))
x = x + [4]                        #连接两个列表
print(x)
print(id(x))                       #内存地址发生改变
x += [5]                           #使用复合赋值运算符为列表追加元素
print(x)
print(id(x))                       #内存地址不变
````

<font face='kaiti' size=5 style='line-height:1.5'>
    
&emsp; 乘法运算符`*`可以用于列表和整数相乘，表示序列重复，返回新列表。运算符`*=`也可以用于列表元素重复，属于原地操作。

````python
x = [1, 2, 3, 4]
print(id(x))
x = x * 2                           #元素重复，返回新列表
print(x)
print(id(x))                        #地址发生改变
x *= 2                              #元素重复，原地进行
print(x)
print(id(x))                        #地址不变
````

<font face='kaiti' size=5 style='line-height:1.5'>

&emsp;&emsp;但`+=`和`*=`用于<font color=#FF0000>数字、字符串、元组等不可变类型数据</font>时，不是原地操作。

````python
# 以下的操作均改变了内存地址
x = 5
print(id(x))
x += 3
print(id(x))
````

````python
t = 'fire!'
print(id(t))
t *= 3
print(id(t))
````

````python
x = (1,2,3)
print(id(x))
x+=(2,3,4)
print(id(x))
````

### 列表的常用方法

<font face='kaiti' size=5 style='line-height:1.8'>
    
&emsp; 这里的方法指的是对象的成员函数，Python中可用内置函数dir()列出列表对象的属性和方法。列表的常用方法能实现列表元素的添加、删除、清空、查找、计数、逆序、排序、浅复制等，见下表所示。

<font size=4>

$$\begin{array}{l|l}\hline
{\bf 方法} & {\bf 描述} \\ \hline
{\rm append()}&向列表末尾追加元素\\

<font color=#0000FF>【例1】</font> 如果一个整数能被9整除，则其各位数字之和也能被9整除。任意输入一个多位整数，用该方法判断其是否能被9整除。

````python
n = input('输入一个整数：')
n = list(n) # 将输入的整数字符串变成列表
alist = []
for k in n: # 将列表中的字符变成数字
    alist.append(int(k))
if sum(alist)%9==0:
    print('能被9整除')
else:
    print('不能被9整除')
````

&emsp;如果一个简单的循环只是为了形成一个列表，则可以尝试用列表推导式代替。列表推导式使用非常简洁的方式来快速生成满足特定需求的列表，代码具有非常强的可读性。<br>

&emsp;设计一个程序，对输入的18位身份证号码辨别其真伪，若为真，则判断性别，若为假，则提示重新输入。

````python
coef = [7,9,10,5,8,4,2,1,6,3,7,9,10,5,8,4,2] # 系数
check = '10X98765432' # 校验码
while True:
    id = input('请输入身份证号码，0则退出 ')
    if id=='0':
        break
    if len(id)!=18:
        print('输入位数不对，请重新输入')        
    else:
        s = sum([coef[i]*int(id[i]) for i in range(17)]) #乘积求和
        code = check[s%11]  # 生成校验码
        if id[-1] == code or id[-1].upper()==code:
            print(id,'为合法身份证号码，',end='')
            if int(id[-2])%2:  #余数非0
                print('男性')
            else:
                print('女性')
        else:
            print(id, '为非法身份证号码') 
````

### 课后阅读：浅复制与深复制

<font face='kaiti' size=5 style='line-height:1.5'>
    
1. 【<font color=#FF5000>列表的浅复制 copy()</font>】<br>
&emsp;&emsp;列表对象的copy()方法返回列表的浅复制。与浅复制相对的，还有Python标准库copy库的深复制deepcopy()。之所以需要浅复制和深复制，是因为Python中的变量并不直接存储值，而是存储了值的内存地址或者引用，当对一个变量做原地操作时，指向同一地址的变量会受影响。见下面的示例：

````python
# 将对象x赋值给y，x重新赋值，不会改变y
x = list('Python')
print(id(x))
y = x           # y与x指向同一地址
print(id(y))
x = x+['3']   # 重新赋值，非原地操作
print(x)
print(id(x))
y
````

````python
# 将对象x赋值给y，对x做原地操作，则y一起改变
x = list('Python')
y = x           # y与x指向同一地址
x.append('3')   # 原地操作，x,y的地址不变，值一起改变
print(x)
print(y)
````

<font face='kaiti' size=5 style='line-height:1.5'>
    
&emsp; 在编程时，我们经常需要改变一个对象的取值，但又需要保留原对象的取值。Python中，将原对象赋值给一个新变量后，两者指向的地址是一致的。如果对其中某一变量重新赋值，这时2个变量将分别指向不同的对象。但如果对某一变量的值做原地操作，则另一个将同样改变，不能保留原来的值。Python中引入浅复制和深复制来解决这个问题。

<font face='kaiti' size=5 style='line-height:1.5'>

&emsp; 所谓浅复制，是指生成一个新的列表，并且把原列表中所有元素的引用都复制到新列表中。如果原列表包含数字、字符串、元组这样的不可变类型的数据，则改变任何一个列表的不可变数据，另一个列表的对应数据不会改变。

````python
x = list('Python')
y = x.copy()
x.append('3')
print(x)
print(y)
````

<font face='kaiti' size=5 style='line-height:1.5'>

2. 【<font color=#FF5000>列表的深复制 copy()</font>】
    
&emsp; 如果原列表的元素有列表之类的可变类型数据，则在任何一个列表中，<font color=#ff0000>对可变类型的元素做原地操作，另一个列表中的对应数据也做同样变化</font>，但如果对可变类型数据做非原地操作，另一个列表中的对应数据不改变。因为Python中，不可变数据不存在原地操作，所以，以上可以概括为：<font color=#ff0000>对列表及其浅复制中的某一个列表的元素做原地操作，则另一列表的对应元素也同样变化</font>。

````python
# 注意下例中对子列表的两种不同操作
x = [1, 2, [3, 4],[5, 6]]
y = x.copy()              # 浅复制
y[2] = y[2] + [5]         # 对新列表中的子列表重新赋值，另一个列表的对应子列表不改变
y.append(6)               # 在新列表尾部追加元素，不影响另一个列表
x[3].append(7)            # 对原列表中的子列表追加数据，另一个列表的子列表同样变化
x[0] = -1                 # 修改不可变类型数据，不影响另一个列表
print(y)
print(x)
````

<font face='kaiti' size=5 style='line-height:1.5'>
    
&emsp; 如果希望列表被复制后，两者元素的原地操作不相互影响，可以使用标准库copy中的<font color=#ff0000>deepcopy()函数实现深复制</font>。这样得到的新列表和原列表是相互独立的，修改任何一个都不会影响另一个。

````python
from copy import deepcopy
x = [1, 2, [3, 4],[5, 6]]
y = copy.deepcopy(x)                           # 深复制，两个列表独立
y[2] = y[2] + [5]                      
y.append(6)                            
x[3].append(7)                         
x[0] = -1                               
print(y)  # 修改y，不影响x
print(x)
````
