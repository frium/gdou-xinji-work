# 5 Python函数的设计

<font face='kaiti' size=5 style='line-height:1.5'> 
    
- 将可能需要反复执行的代码封装为函数，并在需要该功能的地方进行调用，不仅可以实现代码复用，更重要的是可以保证代码的一致性，只需要修改该函数代码则所有调用均发生改变。

## 5.1 函数定义与调用

### 5.1.1 函数的定义

<font face='kaiti' size=5 style='line-height:1.5'>
    
&emsp;Python函数定义语法如下：
    
----
    
```python
def 函数名(参数序列):
    '''（多行）注释'''
    
    函数体
```
---------

<font face='kaiti' size=5 style='line-height:1.8'>
    
&emsp;&emsp;Python中使用关键字**def**来定义函数，然后是空格和函数名，函数名后面的圆括号内是形式参数（formal-parameters简称“形参”）序列。如果有多个参数，则用逗号隔开。

<font face='kaiti' size=5 style='line-height:1.8'>

&emsp;&emsp;注意事项:
- 在定义函数时，开头部分的注释并不是必需的，但适当的注释可以为用户提供友好的提示。
- 定义语句中，不需要声明形参的类型，也不需要指定函数返回值类型。
- 即使该函数不需要接收任何参数，函数名后也必须保留一对空的圆括号。
- 圆括号后面的冒号必不可少。
- 函数体相对于def关键字必须保持一定的空格缩进。

### 5.1.2 函数的调用


<font face='kaiti' size=5 style='line-height:1.5'>

&emsp;函数调用的一般形式如下：
```python
函数名(参数序列)
```

<font face='kaiti' size=5 style='line-height:1.8'>

&emsp;在函数调用时，参数序列中是要传入函数内部的数据，称为实参（actual-parameters）。实参的值会传递给函数内部的形参，在函数体中代替形参进行运算。<br>
&emsp;在Python中，所有的语句都是实时执行的，不像C/C++存在编译过程。def也是一条可执行语句，定义一个函数。所以<font color=#ff0000>函数的定义必须放在函数调用之前</font>。

<font face='kaiti' size=5 style='line-height:1.8'>
    
【<font color=#0000FF>例1</font>】下面的代码是求斐波那契数列中不大于数n的所有值，将其用函数的形式实现，并调用该函数。

````python
# 改写前的代码
n = int(input("请输入斐波那契数列的上界："))
a, b = 1, 1
while a <= n:   # 输出不大于n的序列，注意冒号是缩进的开始
    print(a, end='  ')
    a, b = b, a+b
````

````python
# 用函数实现后的的代码
def fib(n):
    '''计算斐波那契数列中不大于数n的所有值'''

    a, b = 1, 1
    while a <= n:
        print(a, end='  ')
        a, b = b, a+b

n = 1000
fib(n)      # 调用函数fib，1000为实参
````

<font face='kaiti' size=5 style='line-height:1.8'> 

&emsp;这里通过语句`fib(1000)`调用函数`fib()`时，先将实参`1000`传递给函数`fib(n)`的形参`n`，`n`被`1000`代替，再执行函数体，输出结果。

<font face='kaiti' size=5 style='line-height:1.8'> 

&emsp;注意：`fib`的函数体中不需要再使用`input`语句输入`n`的值，这里`n`由函数调用时的实参赋值。

<font face='kaiti' size=5 style='line-height:1.5'> 

&emsp;函数定义时，开头的注释将在帮助信息中显示，如下所示：

````python
help(fib)      # 查看函数fib的帮助信息
````

### 5.1.3 函数的返回

<font face='kaiti' size=5 style='line-height:1.8'>    
    
&emsp;为便于在后续程序中使用函数中的运算产生的数据，需要有返回值。Python中使用**return**语句给出函数的返回值，并退出函数。

<font face='kaiti' size=5 style='line-height:1.8'> 
    
- return可以返回任意类型的值，且不需要声明函数的返回值类型，由return语句返回的表达式的类型确定。
- 不论return语句出现在函数的什么位置，一旦得到执行将直接结束函数的执行。
- 如果函数没有return语句、或者有return语句但是没有得到执行、或者return后不带任何值，解释器都会认为该函数以return None结束，即返回空值。

<font face='kaiti' size=5 style='line-height:1.5'>【<font color=#0000FF>例2</font>】为例1中的函数fib(n)添加返回值，即返回斐波那契数列中不大于数n的所有值：

````python
# 有返回值的函数定义
def fib(n):     
    '''返回斐波那契数列中不大于数n的所有值'''
    if n < 1:
        return    # 返回空值    
    else:
        t = [1, 1]        
        while t[-1]+t[-2] < n:
            t.append(t[-2]+t[-1])
        return t

print(fib(10))
print(fib(0))      # 返回空值
print(fib(1000))
````

<font face='kaiti' size=5 style='line-height:1.5'>
    
&emsp;上述的代码也可以简化如下。其中对 `n<1` 的情况没有定义，将返回空值。

````python
def fib(n):     
    '''返回斐波那契数列中小于数n的所有值'''
    if n>=1:
        t = [1, 1]        
        while t[-1]+t[-2] <= n:
            t.append(t[-2]+t[-1])            
        return t

print(fib(13))
print(fib(1000)) 
print(fib(0))         # 返回空值
````

###  5.1.4 函数嵌套定义

<font face='kaiti' size=5 style='line-height:1.5'>

&emsp;Python允许函数的嵌套定义，在函数内部可以再定义另外一个函数。<br>

<font face='kaiti' size=5>【<font color=#0000FF>例3</font>】编写函数，返回不大于n的所有质数。

````python
def allprime(n):
    '''返回不大于n的所有质数'''
    if n >= 2:
        def isprime(x):
            '''内部函数isprime(x)用于判断x是否为质数'''
            for k in range(2, x//2 + 1):  # 如果x能被k整除，则不是质数
                if x % k == 0:
                    return False
            return True  # 判断为质数

        t = []
        for x in range(2, n+1):
            if isprime(x):
                t.append(x)
        return t

alist=allprime(100)
print(alist)
````

<font face='kaiti' size=5 style='line-height:1.8'>

- 尽管函数嵌套定义使用很方便，也很灵活，但并不提倡过多使用，因为这样会导致内部的函数反复定义而影响执行效率。
- 如果内部函数是其他场合也需要使用的，可以脱离嵌套的形式，在调用它的函数前定义。

<font face='kaiti' size=5 style='line-height:1.5'>
    
&emsp;则上述的代码可以修改如下：

````python
def isprime(x):
    '''判断x是否为质数'''    
    for k in range(2, x//2 + 1):  # 如果x能被k整除，则不是质数
        if x % k == 0:
            return False            
    return True  # 判断为质数

def allprime(n):
    '''返回不大于n的所有质数'''
    if n >= 2:        
        t = []
        for x in range(2, n+1):
            if isprime(x):
                t.append(x)
        return t

alist = allprime(100)
print(alist)
````

## 5.2 函数的参数类型

<font face='kaiti' size=5 style='line-height:1.5'>

&emsp;函数在定义时的输入参数称为形式参数（简称形参），通常没有具体的取值。在函数调用时，给定的输入参数称为实际参数（简称实参）。如上述的斐波拉契函数定义语句 `fib(x)`中的 `x` 为形式参数，而在调用时，`fib(1000)`中的 `1000` 为实际参数。 

<font face='kaiti' size=5 style='line-height:1.5'>

&emsp;为了增加函数调用的灵活性和方便性，Python函数的参数在定义或调用时，可有多种不同的形式，代表不同的含义或作用。

### 5.2.1 位置参数(positional arguments)

<font face='kaiti' size=5 style='line-height:1.5'>

&emsp;位置参数，意味着在定义时按顺序赋予参数的不同意义，而在调用时按相同的顺序输入对应意义的参数。即在函数定义时，形参没有设置默认值，而在函数被调用时也没有使用形参的名称，实参与形参的顺序和数量必须一致。

<font face='kaiti' size=5>【<font color=#0000FF>例4</font>】使用位置参数，定义一个 $\Delta$ 函数.

````python
def delta(a, b, c):
    return b*b - 4*a*c

delta(1,3,2)  # 这里实参1,3,2分别对应到形参a,b,c
````

### 5.2.2 默认值参数(Default Arguments)

<font face='kaiti' size=5 style='line-height:1.8'>

&emsp;在函数定义时，可为形参赋值，该值即为形参的默认值。调用函数时如果未对该形参提供实参，则该实参取默认值。

<font face='kaiti' size=5 style='line-height:1.5'>

  - 在调用含有默认值参数的函数的过程中，实参到形参的传递也是按位置顺序逐个赋值；
  - 定义含默认值参数的函数时，任何一个默认值参数右边都不能再出现没有默认值的普通位置参数，否则会提示语法错误。

<font face='kaiti' size=5 style='line-height:1.5'>
 
【<font color=#0000FF>例5</font>】定义正态分布概率密度函数 $f(x) = \frac1{\sqrt{2\pi}\sigma}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$ 如下，体会默认值参数的作用。

````python
def normpdf(x, mu=0.0, sigma=1.0):
    '''正态分布概率密度函数在点x处的函数值，
    均值mu默认取0.0，标准差sigma默认取1.0.'''

    from math import pi, exp, sqrt
    return exp(-(x-mu)**2/2/sigma**2)/sqrt(2*pi)/sigma


print(normpdf(0.5))
print(normpdf(0.5, 0.0, 1.0))
print(normpdf(1, 2))
print(normpdf(1, 2, 1.0))
````

<font face='kaiti' size=5 style='line-height:1.8'>

&emsp; 如上例所示，在调用函数时，可混合使用普通位置参数和默认值参数的默认值，这时函数会将输入的实参当将普通位置参数，而后面缺少的其他实参则使用其默认值。

### 5.2.3 关键字参数(Keyword Arguments)

<font face='kaiti' size=5 style='line-height:1.8'>
    
&emsp;以上的实参到形参的传递是按位置顺序逐个赋值，这是默认的方式。但当参数很多时，这种调用参数的方式可读性较差，且需记忆各参数的含义及其位置。Python允许在函数调用时对实参指定对应的形参名称，明确指定哪个值传递给哪个形参，这时实参顺序可以任意调整。<br>
&emsp;参数按名称传递时称为<font color=#FF5000>关键字参数</font>(Keyword Arguments)，只需在调用时显式地将实参对应赋值给形参，<font color=#FF5000>并不需要在函数定义时对参数做其他特别设定</font>。

<font face='kaiti' size=5 style='line-height:1.5'>
    
【<font color=#0000FF>例6</font>】对上面定义的正态分布概率密度函数，在调用时使用关键字参数传递参数名称。

````python
def normpdf(x, mu=0.0, sigma=1.0):
    '''正态分布概率密度函数在点x处的函数值，
    均值mu默认取0.0，标准差sigma默认取1.0.'''

    from math import pi, exp, sqrt
    return exp(-(x - mu)**2 / 2 / sigma**2) / sqrt(2 * pi) / sigma


print(normpdf(mu=2, x=1.5, sigma=2.0))
````

<font face='kaiti' size=5 style='line-height:1.5'>
    
&emsp;&emsp;在函数调用时，可以混合使用使用位置参数，关键字参数，以及默认值参数的默认值，这时应将普通位置参数放在前面，关键字参数放在后面，函数会自动对缺少的其他实参使用其默认值。

````python
# 关键字参数应放在位置参数后面，缺少的实参使用默认值
print(normpdf(1.5, 2, 1))
print(normpdf(1.5, mu=2))  # 缺少实参sigma
print(normpdf(mu=2, x=1.5))
print(normpdf(2, 0, 1.5))
print(normpdf(2, sigma=1.5))  # 缺少实参mu
print(normpdf(sigma=1.5, x=2))
````

<font face='kaiti' size=5 style='line-height:1.8'>

&emsp;在函数调用时，普通位置参数不能出现在关键字参数的右边。以下的调用方式将出错。

````python
print(normpdf(1, mu=2, 3.0))  # 普通位置参数不能出现在关键字参数右边
````

<font face='kaiti' size=5 style='line-height:1.8'>
    
&emsp;&emsp;为了避免普通位置参数和关键字参数的位置混乱，也可在函数定义时用"\*"强制将位置参数和关键字参数隔开。如math标准库的isclose函数的定义如下：
    
---
```python
math.isclose(a, b, *, rel_tol=1e-09, abs_tol=0.0)
```
---
这里形参列表中有一个"\*"，这意味着"\*"后面的参数在函数调用时如果出现的话，则必须当做关键字参数使用。

````python
import math
x = 3.14159265
math.isclose(x, math.pi, rel_tol=1e-8)
````

<font face='kaiti' size=5>以下的调用方式中"\*"后面不是关键字参数形式，将出错。

````python
import math
x = 3.14159265
math.isclose(x, math.pi, 1e-8)
````

<font face='kaiti' size=5 style='line-height:1.8'>
    
【<font color=#0000FF>例7</font>】重新定义上面的正态分布概率密度函数，强制要求均值和方差使用关键字参数。

````python
def normpdf(x, *, mu=0.0, sigma=1.0):
    '''正态分布概率密度函数在点x处的函数值，
    均值mu默认取0.0，标准差sigma默认取1.0.'''

    from math import pi, exp, sqrt
    return exp(-(x - mu)**2 / 2 / sigma**2) / sqrt(2 * pi) / sigma

print(normpdf(1.5, mu=2))
print(normpdf(mu=2, x=1.5, sigma=1.0))
````

````python
# 按上面的函数定义，只能使用1个位置参数
print(normpdf(1.5, 2))  
````

### 5.2.4 序列和字典作为实参

<font face='kaiti' size=5 style='line-height:1.8'>
    
&emsp;函数定义中往往有多个形参，而调用函数时的实参可能来自序列（如列表、元组）或字典，Python中允许直接使用序列或字典作为实参。 

<font face='kaiti' size=5 style='line-height:1.8'>
 
&emsp;如果实参来自序列时，函数调用时可以在实参序列前加一个星号`*`将其解包，然后传递给多个单变量形参。

<font face='kaiti' size=5 style='line-height:1.5'>
    
【<font color=#0000FF>例8</font>】调用正态分布概率密度函数，对实参使用序列解包

## 5.4 递归函数<br>
<font face='kaiti' size=5 style='line-height:1.5'>

&emsp;&emsp;函数的递归定义是允许在函数定义内部，函数调用自己，自己再调用自己，自己再调用自己，...，当某个条件得到满足的时候就不再调用了，然后再一层一层地返回直到该函数第一次调用的位置。常用于迭代型函数的定义。



<font face='kaiti' size=5 style='line-height:1.5'>

【<font color=#0000FF>例10</font>】用递归函数形式定义一个函数fib(n)，返回斐波拉契数列的第n项。

````python
def fib1(n):
    '''使用递归法求斐波那契数列第n项的值。'''
    if n in {1,2}:
        return 1    
    else:
        return fib1(n - 2) + fib1(n - 1)

fib1(10)
````

<font face='kaiti' size=5 style='line-height:1.5'>

【<font color=#0000FF>例11</font>】用递归函数形式定义一个函数fib(n)，返回斐波拉契数列的前n项。

````python
def fib2(n):
    '''使用递归法求斐波那契数列前n项的值。'''
    if n == 1:
        return [1]
    elif n == 2:        
        return [1,1]
    else:
        t1 = fib2(n - 1)
        return t1 + [t1[-1] + t1[-2]]  # 这里需要返回列表，不能用原地操作append

fib2(10)
````

<font face='kaiti' size=5 style='line-height:1.8'>

&emsp;&emsp;每次调用函数时需记住离开的位置才能保证函数运行结束时回到正确的位置，这个过程称为保存现场，需要一定的栈空间。另外，调用一个函数时会为其分配一个栈帧，在调用结束自动释放。而函数递归调用时，一个函数执行尚未结束就又调用自己，原来的栈帧尚未释放又分配了新的栈帧，需占用大量的栈空间。所以，递归深度如果太大，可能会导致空间不足进而程序崩溃。

## 【<font color=#FF5000>5.1 实参与形参的关系（选学）</font>】

<font face='kaiti' size=5 style='line-height:1.8'>
    
&emsp;&emsp;在Python中，变量保存的是对象的地址，类似C/C++中的指针。在调用函数时，实参传递给形参的是实参的引用而非实参的值。<br>
&emsp;&emsp;一般来说，在函数内部对形参的修改不会影响实参。但如果参数为可变数据类型，在函数内部对可变类型的形参的元素做了增、删、改的操作，则同样的作用会体现在实参上。

````python
def minu(v):  # 使用下标修改列表元素值为其相反数
    for k in range(len(v)):
        v[k] = -v[k]
    return v


a = [2, 3, 4]
v = minu(a)
print(v)
a #可以发现，a的值随v一起改变了
````

````python
def subjoin(v, item):  # 使用列表的方法为列表添加加元素
    v.append(item)
    return v


a = [2, 3, 4]
print(subjoin(a, 5))
a #可以发现，a的值随v一起改变了
````

<font face='kaiti' size=5 style='line-height:1.8'>

&emsp; 上述的例子中，在函数中对形参的操作，实现了对可变类型实参的修改。如果要避免可变类型的实参被函数修改，可以在定义函数时函数内部使用形参的浅复制或深复制，或在调用函数时使用可变类型实参的浅复制或深复制作为实参。

````python
def subjoin(v, item):    # 使用列表的方法为列表增加元素
    x = v.copy()
    x.append(item)
    return x

a = [2,3,4]
print(subjoin(a,5))
a # a没有随之改变
````

````python
def normpdf(x, *p):
    '''正态分布概率密度函数在点x处的函数值，
    *p代表2个参数：均值mu默认取0.0，标准差sigma默认取1.0.'''

    def parse(p):  # 内部函数，用于处理输入的可变参数*p
        mu = 0.0
        sigma = 1.0
        t = len(p)
        if t == 0:
            return (mu, sigma)
        elif t == 1:
            return (p[0], sigma)
        else:  # 如p的长度>=2，则原样返回
            return (p)

    args = parse(p)  # 先调用内部函数，获得mu和sigma
    if len(args) > 2:
        print('错误，可变参数"*p"最多包含2个参数mu和sigma!')
    else:
        (mu, sigma) = parse(p)
        from math import pi, exp, sqrt
        return exp(-(x - mu)**2 / 2 / sigma**2) / sqrt(2 * pi) / sigma


print(normpdf(1.5, 2))
print(normpdf(1.5))
print(normpdf(1.5, 2, 1.0))
print(normpdf(1.5, 2, 3, 1))  # 可变参数*p包含的参数多于指定的个数
````

## 【<font color=#FF5000>5.3 变量作用域（选学）</font>】

<font face='kaiti' size=5 style='line-height:1.5'>
    
&emsp; 变量起作用的代码范围称为变量的作用域，不同作用域内变量名可以相同，互不影响。在函数内部定义的普通变量只在函数内部起作用，称为局部变量。在函数外部定义的变量为全局变量。局部变量的引用比全局变量速度快，应优先考虑使用。<p>
&emsp; 如果函数内的局部变量与函数外的全局变量具有相同的名字，那么该局部变量会在自己的作用域内隐藏同名的全局变量。<p>
&emsp; 函数内定义的局部变量只在该函数内可见，当函数执行结束后，局部变量自动删除，不再可以使用。而在函数内使用global定义的全局变量当函数调用结束后仍然存在且可以访问。

<font face='kaiti' size=5 style='line-height:1.8'>

全局变量可以通过关键字global来定义。这分为两种情况：
1. 一个变量已在函数外定义，如果在函数内需要为这个变量赋值，并要将这个赋值结果传递到函数外，可以在函数内使用global将其声明为全局变量。
2. 如果一个变量在函数外没有定义，在函数内部也可以直接将一个变量定义为全局变量，该函数执行后，将增加一个新的全局变量。

<font face='kaiti' size=5 style='line-height:1.5'>
    
【<font color=#0000FF>例14</font>】在上述的正态分布概率密度函数的内部函数中，使用global定义的全局变量传递均值和方差。

````python
def normpdf(x, *p):
    '''正态分布概率密度函数在点x处的函数值，
    *p代表2个参数：均值mu默认取0.0，标准差sigma默认取1.0.'''

    def parse(p):  # 内部函数，用于处理输入的可变参数*p
        global args  # 函数没有返回，通过全局变量args传递参数
        t = len(p)
        if t == 0:
            args = [0.0, 1.0]
        elif t == 1:
            args = [p[0], 1.0]
        else:  # 如p的长度>=2，则原样返回
            args = p

    parse(p)
    if len(args) > 2:
        print('错误，可变参数"*p"最多包含2个参数mu和sigma!')
    else:
        (mu, sigma) = args
        from math import pi, exp, sqrt
        return exp(-(x - mu)**2 / 2 / sigma**2) / sqrt(2 * pi) / sigma


print(normpdf(1.5, 2))
print(normpdf(1.5))
print(normpdf(1.5, 2, 1.0))
````
