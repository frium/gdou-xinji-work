二、（15分）期末考试成绩按百分制分成5档，90分及以上为优秀，80-89分为良好，70-79分为中等，60-69分为及格，低于60分为不及格。编写一个函数grade(score)，可对输入的成绩score，输出其对应的等级。若分数在[0,100]之外，则输出为“成绩错误”。

```python
# 以下为函数定义

def grade(score):
    if score>100 or score<0:
        return '成绩错误'
    elif score>=90:
        return '优秀'
    elif score>=80:
        return '良好'
    elif score>=70:
        return '中等'
    elif score>=60:
        return '及格'
    else:
        return '不及格'
    
# 以下为函数调用
print(grade(71))
print(grade(43))
print(grade(94))
print(grade(105))
```

三、（10分）求平方根 $\sqrt{C}$ 的一种迭代公式为 $x_{k+1}=\dfrac12(3x_k-\dfrac{x_k^3}{C}).$ 利用该公式求 $\sqrt{3}$。迭代初始值可以取1，当满足 $|x_{k+1}-x_k|<10^{-8}$ 时，停止迭代，并取$x_{k+1}$为平方根的近似值. 

```python
def mySqrt(C, tol=1e-6):
    x0 = 1
    while True:      # while后面取 True 的作用是在循环前不需迭代1次，相当于do...while循环
        x1 = (3*x0-x0**3/C)/2
        if abs(x1-x0)<tol:
            return x1 
        x0 = x1

mySqrt(3,1e-8)
```

```python
x0, C = 1, 3
x1 = (3*x0-x0**3/C)/2
while abs(x1-x0) > 1e-8:  # 注意while下方的2行语句的先后顺序
    x0 = x1
    x1 = (3*x0-x0**3/C)/2
x1
```

四、(20分) 分别绘制下列图形：<br>
1. 某同学去年4个季度的生活费分别为7000, 7300, 6100, 7500元。在大小为(8,4)的图形窗口中，分别在左右子图中用条形图和饼图表现这些数据；条形图横坐标为['Spr.','Sum.','Fal.','Win.']，标题为'expenses'；饼图上应显示比例，标题为'ratio'。<br>
2. 曲面$z = -\dfrac{x^3}8+\dfrac{y^2}4$在区域 $x,y\in[-3,3]$上的图形。提示：在`add_subplot()`中添加参数`projection='3d'`即可创建3维坐标系ax，用`ax.plot_surface()`可绘制曲面图。

```python
# 1. 条形图和饼图
import numpy as np
import matplotlib.pyplot as plt

cost = [7000, 7300, 6100, 7500]
label = ['Spr.', 'Sum.', 'Fal.', 'Win.']
plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.bar(label, cost)
plt.title('expenses')
plt.subplot(122)
plt.pie(cost, labels=label, autopct="%.1f%%")
plt.title('ratio')
plt.show()
```

```python
# 2. 曲面图
t = np.linspace(-3, 3, 101)
x, y = np.meshgrid(t, t)
z = -x**3 / 8 + y * y / 4
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='jet')
plt.show()
```
