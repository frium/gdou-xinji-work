二、(15分) 某停车场按停车时间（单位：小时）收费，标准如下：

- 小车：(0,1]: 0元；(1, 2]: 5元；(2,4]: 8元；(4,12]: 15元；(12,24]: 20元。
- 大车：(0,1]: 0元；(1, 2]: 8元；(2,4]: 12元；(4,12]: 20元；(12,24]: 30元。
                                                  
编写一个函数parking_fee(time, motor)，实现这一收费方案，对输入的停车时间time和车型motor（0为小车，1为大车，默认为小车），返回其对应费用。

```python
# 以下为函数定义
def parking_fee(time, motor=0):
    if motor==0:
        if time<=1:
            cost = 0
        elif time<=2:
            cost = 5
        elif time<=4:
            cost = 8
        elif time<12:
            cost = 15
        elif time<24:
            cost = 20
    elif motor ==1:
        if time<=1:
            cost = 0
        elif time<=2:
            cost = 8
        elif time<=4:
            cost = 12
        elif time<12:
            cost = 20
        elif time<24:
            cost = 30
    return cost


# 以下为函数调用
print(parking_fee(3,1))
print(parking_fee(5))
print(parking_fee(10,0))
print(parking_fee(21,1))
```

```python
# 来自信计1234吴湘帆同学，代码比较简短
# 以下为函数定义
def parking_fee(time: int,motor=0) -> int:
    fee: list = [0,5,8,15,20]
    if motor:
        fee = [0,8,12,20,30]
    time_list = [1,2,4,12,24]
    for index,t in enumerate(time_list):
        if time <= t:
            return fee[index]
        
# 以下为函数调用
print(parking_fee(3,1))
print(parking_fee(5))
print(parking_fee(10,0))
print(parking_fee(21,1))
```

```python
# 下面将吴湘帆同学的函数定义改得更简短一点，不过程序的可读性要差一点
# 以下为函数定义
def parking_fee(time, motor=0):
    bins = [0, 1, 2, 4, 12, 24]
    cost = [[0, 5, 8, 15, 20],[0, 8, 12, 20, 30]] # 二维数组
    return cost[motor][sum([time>=x for x in bins])-1]

# 以下为函数调用
print(parking_fee(3,1))
print(parking_fee(5))
print(parking_fee(10,0))
print(parking_fee(21,1))
```

三、（10分）一个数恰好等于它的平方数的右端，这个数称为“同构数”。如6是其平方数36的右端，25是其平方数625的右端，所以，6和25都是同构数。求[2,1000]上所有的同构数。

```python
tonggou = []
for k in range(2,1001):
    ksq = k*k
    d = len(str(k)) # 利用len(str(k))求k的位数
    if (ksq-k)%(10**d)==0:
        tonggou.append(k)

print('[2,1000]上所有的同构数为：',tonggou)
```

```python
# 不使用str转换为字符串，直接从后往前一一比对
tonggou = []
for k in range(2,1001):
    k_copy = k # 因迭代中k值改变，故先保留k
    ksq = k*k    
    while k>0:
        k_end, ksq_end = k%10, ksq%10 # 末位数
        k, ksq = k//10, ksq//10 # 去除末位数
        if k_end-ksq_end:  # 末位数不同
            break
    else:
        tonggou.append(k_copy)
        
print('[2,1000]上所有的同构数为：',tonggou)
```

```python
# 不使用str提取位数，而通过不断整除10求位数
def wei(x):
    n = 0
    while x>0:
        x=x//10
        n+=1
    return n

tonggou = []
for k in range(2,1000):
    ksq = k*k
    if (ksq-k)%(10**wei(k))==0:
        tonggou.append(k)
tonggou
```

```python
# 不使用str提取位数，而通过与10的乘方比较大小求位数
def wei(x):
    n = 0
    while x>=10**n:        
        n+=1
    return n

tonggou = []
for k in range(2,1000):
    ksq = k*k
    if (ksq-k)%(10**wei(k))==0:
        tonggou.append(k)
tonggou
```

```python
# 利用字符串的endswith方法
tonggou = []
for k in range(2,1001):
    ksq = str(k*k)
    if ksq.endswith(str(k)):
        tonggou.append(k)
        
print('[2,1000]上所有的同构数为：',tonggou)
```

```python
# 建立一个函数，判断是否为同构数
def tg(num):
    sq = str(num**2)
    return str(num)==sq[-len(str(num)):]

tonggou =[num for num in range(2,1001) if tg(num)]
print('[2,1000]上所有的同构数为：',tonggou)
```

四、(20分) 分别绘制下列图形：<br>
1. 某公司今年1~6月的营业额为3040, 3577, 4322, 4210, 5076, 4875万元。分左右子图，用条形图和饼图表现这些数据及其比例。条形图的横坐标为对应月份，标题为“income”；饼图上显示的百分比精确到小数点后2位，标题为'ratio'。<br>
2. 曲面$z = \dfrac{x^2}2+\dfrac{y^3}5$在区域 $x,y\in[-2,2]$上的图形。提示：在`add_subplot()`中添加参数`projection='3d'`即可创建3维坐标系ax，用`ax.plot_surface()`可绘制曲面图。

```python
# 1. 条形图和饼图
import numpy as np
import matplotlib.pyplot as plt

cost = [3040, 3577, 4322, 4210, 5076, 4875]
plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.bar(range(1,7), cost)
plt.xticks(range(1,7))
plt.title('income')
plt.subplot(122)
plt.pie(cost, labels=range(1,7), autopct="%.2f%%")
plt.title('ratio')
plt.show()
```

```python
# 2. 曲面图
import numpy as np

t = np.linspace(-2, 2, 101)
x, y = np.meshgrid(t, t)
z = x * x / 2 + y**3 / 5
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='jet')
plt.show()
```

（1）建立一个名为'temp'的DataFrame，存放上述数据，列标签分别为'city', 'Feb', 'Mar'.<br>
（2）在上述DataFrame后插入两月平均气温的平均值，列标签为'avg'。又因为2月为28天，而3月为31天，在上述DataFrame后再插入一列平均气温数据，该平均值是考虑各月天数时的平均气温，列标签为'wavg'.<br>
